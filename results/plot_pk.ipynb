```python
import numpy as np
import matplotlib
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
rc_fonts = {
    "font.family": "serif",
    "font.size": 20,
    'figure.figsize': (5, 3),
    "text.usetex": True,
}
matplotlib.rcParams.update(rc_fonts)
from matplotlib import pyplot as plt
```

```python
k_pre,pk_pre=np.loadtxt('pre-recon_pk_iterattive.txt',usecols=(0,1),unpack=True)
k_post_iter,pk_post_iter=np.loadtxt('post-recon_pk_iterattive.txt',usecols=(0,1),unpack=True)
k_post_multi,pk_post_multi=np.loadtxt('post-recon_pk_multigrid.txt',usecols=(0,1),unpack=True)
```

```python
fig,ax=plt.subplots()
ax.plot(k_pre,k_pre*pk_pre,label='pre-recon',ls='-',c='#bb5566')
ax.plot(k_post_multi,k_post_multi*pk_post_multi,ls='-.',label='post-recon, multigrid',c='#004488')
ax.plot(k_post_iter,k_post_iter*pk_post_iter,ls=':',label='pre-recon, iterattive',c='#ddaa33')
ax.set_ylim(-50,1600)
ax.set_ylabel('$kP(k)$')
ax.set_xlabel('$k$')
ax.legend(frameon=False,loc='lower left')
```
